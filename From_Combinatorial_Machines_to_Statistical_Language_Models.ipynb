## Pre-20th Century to Mid-20th Century: From Combinatorial Machines to Statistical Language Models

### Pre-20th Century: Combinatorics and Chance Composition

**c. 1274–1316 — Ramon Llull, *Ars Magna***  
Developed a mechanical and combinatorial method for generating propositions by rotating labeled figures.  
This represents one of the earliest **algorithmic generative systems**, anticipating formal rule-based generation centuries before computers.  
*Sources:* Stanford Encyclopedia of Philosophy, Wikipedia  

**18th Century — “Musical Dice Games” (Musikalisches Würfelspiel)**  
Composers assembled music by rolling dice to select from prewritten musical fragments.  
Associated with **Kirnberger (1757)**, **C. P. E. Bach (1758)**, and later **works attributed to Mozart (1780s–1790s)**.  
An early example of *stochastic composition* combining randomness with structural rules.  
*Sources:* Wikipedia  

---

### 19th Century: Computing Imagined as a Generator

**1843 — Ada Lovelace, Notes on Babbage’s Analytical Engine**  
Anticipated algorithmic creativity, proposing that “the engine might compose… music of any degree of complexity.”  
This was the **conceptual root of machine generativity**, connecting computation with artistic creation.  
*Sources:* Classical Music  

---

### Early 20th Century: Statistical Sequence Generation

**1913 — Andrey Markov**  
Applied *Markov chains* to analyze the vowel–consonant structure in Pushkin’s *Eugene Onegin*.  
This inaugurated the mathematical modeling of **probabilistic sequence generation**, demonstrating that symbol prediction could emerge from statistical dependencies.  
*Sources:* ALPHA 60, Aurora  

---

### Mid-20th Century: Information Theory and Early Computer Outputs

**1948 — Claude E. Shannon, “A Mathematical Theory of Communication”**  
Introduced **entropy**, **mutual information**, and the **n-gram model** concept.  
By generating English-like text from **0-, 1-, 2-, and 3-gram** statistics, Shannon presented the **first formally defined probabilistic text generator**.  
*Sources:* Harvard Mathematics, Web Archive  

**1952–1953 — Christopher Strachey, *Love-Letter Generator***  
Developed on the **Manchester Mark 1** computer, this program used template structures and randomized lexicon selection to produce human-like love letters.  
One of the earliest **computer-generated literary works**.  
*Sources:* Wikipedia, The New Yorker  

**1957 — Hiller & Isaacson, *Illiac Suite***  
The first complete **computer-composed musical score**, built using stochastic processes, probabilistic rules, and early grammar-based algorithms.  
Bridged algorithmic art and computational creativity.  
*Sources:* Wikipedia, Department of Music  

---

### 1960s–1970s: Rule-Based Generative Text and Art

**1966 — Joseph Weizenbaum, *ELIZA***  
A rule-based system that generated conversational text through pattern matching and response templates.  
A milestone in **dialogue generation** and early AI linguistics.  
*Sources:* ACM Digital Library  

**1960s — Cut-up Techniques (Tzara → Gysin/Burroughs)**  
Revived the Dadaist cut-up method—physically rearranging text fragments to create new compositions.  
Served as a **cultural precursor to stochastic and generative text synthesis**.  
*Sources:* Center for Contemporary Writing Programs, Take Dada Seriously  

**1973 onward — Harold Cohen’s *AARON***  
A pioneering **rule-based visual art system**, producing autonomous drawings and paintings that evolved over decades.  
A milestone in computer-based generative art and creative AI.  
*Sources:* Computer History Museum, Whitney Museum  

---

### Where Shannon Fits

Shannon’s 1948 demonstrations form the **first mathematically formalized generative text model**.  
They stand on a historical continuum beginning with Llull’s mechanical combinatorics, through 18th-century stochastic music, Markov’s 1913 sequence statistics, and culminating in computer-based text generation.  
*Sources:* Stanford Encyclopedia of Philosophy, Harvard Mathematics, ALPHA 60  

---

### TL;DR Summary

- **Medieval and Early Modern Roots:** Llull’s combinatorial engines (13th c.) and musical dice games (18th c.) represent early *generative systems*.  
- **Statistical Generation:** Markov (1913) initiates probabilistic sequence modeling; Shannon (1948) formalizes it with information theory.  
- **Early Computational Generativity:** Strachey’s *Love-Letter Generator* (1952–53) and Hiller & Isaacson’s *Illiac Suite* (1957) are the **first computer outputs**.  
- **Rule-Based AI Emergence:** Weizenbaum’s *ELIZA* (1966) and Cohen’s *AARON* (1973→) establish **AI-driven generative text and art**.
