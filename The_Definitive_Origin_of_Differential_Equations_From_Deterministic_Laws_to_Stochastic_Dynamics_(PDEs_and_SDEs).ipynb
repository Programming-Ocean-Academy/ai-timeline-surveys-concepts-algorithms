## The Definitive Origin of Differential Equations: From Deterministic Laws to Stochastic Dynamics

# Who First Invented PDEs and SDEs?  
## A Precise Historical and Scientific Answer

This answer distinguishes **conceptual origins**, **physical formulations**, and **rigorous mathematical invention**, which is essential for historical accuracy.

---

## Who First Invented PDEs (Partial Differential Equations)?

### 1. Conceptual Origin: Calculus as the Enabler  
**Late 17th century**

- **Isaac Newton**
- **Gottfried Wilhelm Leibniz**

Newton and Leibniz independently invented **differential and integral calculus**, providing the mathematical language of derivatives and integrals.

**Role in PDE history**

- Made partial derivatives mathematically possible  
- Enabled multivariable analysis  
- Did *not* formulate PDEs as governing laws of systems  

Thus, calculus is a **necessary prerequisite**, but not the invention of PDEs as a discipline.

---

### 2. First Systematic Formulation of PDEs  
**1822 — Joseph Fourier**

**Work**
- *Théorie analytique de la chaleur*

**Key achievements**

Fourier introduced the **heat equation**:

$$
\frac{\partial u(x,t)}{\partial t} = \kappa \nabla^2 u(x,t),
$$

and, crucially, established:

- PDEs as **fundamental laws governing continuous physical systems**
- Initial and boundary value problems as central objects of study
- Fourier series as a systematic solution method
- The idea that macroscopic physical behavior follows PDEs

**Why Fourier is the correct answer**

Fourier was the first to:
- Derive a PDE directly from physical principles
- Treat PDEs as universal dynamical laws
- Build a full analytical framework around them

---

### Correct Historical Answer (PDEs)

- **Calculus (enabler):** Newton & Leibniz  
- **PDEs as a scientific discipline:** **Joseph Fourier (1822)**

---

## Who First Invented SDEs (Stochastic Differential Equations)?

### 1. Conceptual Origin: Random Motion in Physics  
**1905–1906**

- **Albert Einstein (1905)**
- **Marian von Smoluchowski (1906)**

They explained **Brownian motion** statistically, showing that microscopic randomness leads to macroscopic diffusion.

**Limitation**
- Motion was described probabilistically, not as differential equations driven by noise
- No pathwise stochastic dynamics were defined

---

### 2. First Stochastic Dynamical Equation (Informal)  
**1908 — Paul Langevin**

**Langevin equation**

$$
dx = f(x)\,dt + \sqrt{2D}\, \xi(t),
$$

where \( \xi(t) \) represents random forcing.

**Key contribution**
- Introduced the idea of **dynamics = deterministic force + noise**
- Established non-equilibrium stochastic motion at the equation-of-motion level

**Limitation**
- Noise term was heuristic
- No rigorous definition of integration with respect to randomness

---

### 3. First Rigorous Mathematical Invention of SDEs  
**1944–1951 — Kiyosi Itô**

**Major contributions**

- **Itô stochastic integral** (1944)
- **Itô calculus**
- Rigorous theory of stochastic differential equations

Formalized equations of the form:

$$
dX_t = f(X_t,t)\,dt + g(X_t,t)\,dW_t,
$$

where \( W_t \) is Brownian motion.

**Why Itô is the inventor of SDEs**

Itô:
- Defined integration with respect to Brownian motion
- Resolved the non-differentiability of noise
- Made stochastic dynamics mathematically precise
- Enabled existence, uniqueness, and calculus for SDEs

Without Itô, SDEs are **not mathematically well-defined objects**.

---

### Correct Historical Answer (SDEs)

- **Conceptual SDE (physics):** Paul Langevin (1908)  
- **Rigorous mathematical SDE:** **Kiyosi Itô (1944–1951)**

---

## One-Line Authoritative Summary

**PDEs:** Enabled by Newton and Leibniz → formalized as physical laws by **Joseph Fourier**.  

**SDEs:** Motivated by Einstein and Langevin → rigorously invented by **Kiyosi Itô**.

---

This distinction—*conceptual motivation vs. mathematical invention*—is essential in both physics and modern AI diffusion theory.
